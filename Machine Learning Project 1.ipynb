# 

Splitting the Features and Target Variable 

```python
#Dropping the Target Column

X = df.drop(columns = 'target', axis = 1)

Y = df['target']
```

```python
#Checking to see if the Target column is dropped properly.

X
```

```python
#Splitting the data into train and test data 

from sklearn.model_selection import train_test_split

X_train, X_test ,Y_train, Y_test = train_test_split (X, Y , train_size = 0.7 , test_size = 0.3, random_state = 2 )
```

```python
print (X.shape, X_train.shape, X_test.shape)
```

# Model training 

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train, Y_train)
```

# Model Evaluation

```python
#Checking the accuracy score on training data 

from sklearn.metrics import accuracy_score

X_train_prediction = model.predict(X_train)

training_accuracy = accuracy_score(X_train_prediction, Y_train)



```

```python
#Checking the accuracy on training data 

print ('Accuracy on training data :', training_accuracy)
```

# Our training data accuracy score is 85%

```python
X_test_prediction = model.predict(X_test)

test_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
#Checking the accuracy score on test data 

print('Accuracy score on test data :', test_accuracy)
```

# Our test data accuracy score is 84%

# Observation: Our training data and test data accuracy score is close to each other which means there is no Overfitting and Underfitting in our model.

# 

To check our model is predicting correct we will make a predictive system.

```python
input_data = (54,1,0,122,286,0,0,116,1,3.2,1,2,2)

input_data_as_numpy_array = np.asarray(input_data)

input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

prediction = model.predict(input_data_reshaped)

prediction

# Also making a conditional statement to check 

if prediction == 0:
    print ('The Person does not have a Heart disease')
    
else:
    print ('The Person has a Heart Disease')    
```

# 0 means that the person does not have a Heart Disease
